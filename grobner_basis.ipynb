```python
from sympy import symbols, groebner
```

```python
# General case

# Define the variables and ordering
t, x, y = symbols('t x y') 
order =[t, x, y]

# Define the polynomials
f1 = t**3 + t*x + y
f2 = 3*t**2 + x

# Create a list of polynomials
polynomials = [f1, f2]

# Compute the Gröbner basis with respect to the specified order
gb = groebner(polynomials, t, x, y, order='lex')

# Print the Gröbner basis
for poly in gb:
    print(poly)
```

```python
# Define the variables and ordering
t, x, y = symbols('t x y') 

# Define the polynomials
input = ["4*t^6-4*y*t^5+y^2*t^4+7*t^4-8*y*t^3+2*y^2*t^2+2*t^2-4*y*t+y^2+x^2-1",
"24*t^5-20*y*t^4+4*y^2*t^3+28*t^3-24*y*t^2+4*y^2*t+4*t-4*y"]
for inputs in input:
    polynomials.append(inputs.replace("^", "**"))

# Compute the Gröbner basis
gb = groebner(polynomials, t, x, y)

# Print the Gröbner basis
print("Gröbner Basis in Python:")
i = 0
for poly in gb:
    i +=1
    print("g_" + str(i) + " = " + str(poly))

print("Gröbner Basis in Latex:")
i = 0
for poly in gb:
    repair = str(poly)
    poly_string = repair.replace("**", "^")
    poly_string = poly_string.replace("*","")
    i +=1
    print("g_" + str(i) + " = " + poly_string)
```
