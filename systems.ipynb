```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
t_min = -1
t_max = 1
n = 30

axlim = 2
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

for t in np.linspace(t_min, t_max, n):
    Fxy =  Y - 2*X*t - t**2
    plt.contour(X, Y, Fxy, levels=[0], colors='black')

Exy = Y + X**2
plt.contour(X, Y, Exy, levels=[0], colors='magenta')

plt.gca().set_aspect('equal')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title(r'$F(x, y, t) = y - 2xt - t^2$')
plt.show()
```

```python
t_min = 0
t_max = 2*math.pi
n = 30
axlim = 3
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

p = 1
for t in np.linspace(t_min,t_max,n):
    Fxy = X * np.cos(t) + Y * np.sin(t) - p
    plt.contour(X, Y, Fxy, levels=[0], colors='black')

Exy = X**2 + Y**2 - p**2
plt.contour(X, Y, Exy, levels=[0], colors='magenta')

plt.gca().set_aspect('equal')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title(r'$F(x, y, t) = x \cos t + y \sin t - p$')
plt.show()
```

```python
t_min = -5
t_max = 5
n = 50
S = 1

axlim = 10
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

for t in np.linspace(t_min, t_max, n): 
    Fxy = X / t + t*Y/2*S  - 1
    plt.contour(X, Y, Fxy, levels=[0], colors='black')

Exy = (2*X*Y / S) - 1
plt.contour(X, Y, Exy, levels=[0], colors='magenta')
plt.gca().set_aspect('equal')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title(r'$F(x, y, t) = \dfrac{x}{t} + \dfrac{ty}{2S} - 1$')
plt.show()
```

```python
t_min = 0.01
t_max = 0.99
n = 15
axlim = 1.2
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

for t in np.linspace(t_min,t_max,n):
    Fxy = X**2 / t**2 + Y**2 / (1 - t**2) - 1
    plt.contour(X, Y, Fxy, levels=[0], colors='black') 

Exy = np.abs(X) + np.abs(Y) - 1
plt.contour(X, Y, Exy, levels=[0], colors='magenta')

plt.gca().set_aspect('equal')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title(r'$F(x, y, t) = \frac{x^2}{t^2} + \frac{y^2}{1-t^2} - 1$')
plt.show()
```

```python
t_min = 0.01
t_max = 0.99
n = 15
S = 1
axlim = 3
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

for t in np.linspace(t_min, t_max, n): 
    Fxy = X**2 / t**2 + Y**2 / (S / (math.pi * t))**2 - 1
    plt.contour(X, Y, Fxy, levels=[0], colors='black')

Exy = S / (2 * math.pi * np.abs(X) * np.abs(Y)) - 1
plt.contour(X, Y, Exy, levels=[0], colors='magenta')

plt.gca().set_aspect('equal')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title(r'$F(x, y, t) = \frac{x^2}{t^2} + \frac{y^2}{\left(\frac{S}{\pi t}\right)^2} - 1$')
plt.show()

```
